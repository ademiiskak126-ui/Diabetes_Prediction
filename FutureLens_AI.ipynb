```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = pd.DataFrame({
    "GRE": np.random.randint(280, 340, 500),
    "TOEFL": np.random.randint(80, 120, 500),
    "Uni_Rating": np.random.randint(1, 5, 500),
    "SOP": np.random.uniform(1, 5, 500),
    "LOR": np.random.uniform(1, 5, 500),
    "CGPA": np.random.uniform(6, 10, 500),
    "Research": np.random.randint(0, 2, 500)
})

# fake but realistic target (Chance of Admit)
data["Chance"] = (
    data["GRE"]/340 +
    data["TOEFL"]/120 +
    data["CGPA"]/10 +
    data["Research"]*0.2
) / 4

data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

X = data.drop("Chance", axis=1)
y = data["Chance"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor(n_estimators=200)
model.fit(X_train, y_train)
```

```python
def predict_all(user_input):
    pred = model.predict([user_input])[0]

    print("\n🎯 Your Admission Chance:", round(pred*100, 2), "%")

    # Explainable AI
    print("\n🧠 Why this result?")
    if user_input[0] < 300:
        print("- Low GRE score reduces chance")
    if user_input[1] < 100:
        print("- TOEFL is below competitive level")
    if user_input[5] < 8:
        print("- CGPA needs improvement")
    if user_input[6] == 0:
        print("- No research experience is a disadvantage")

    # Roadmap
    print("\n🧭 Personalized roadmap:")
    if user_input[1] < 100:
        print("👉 Improve English (TOEFL 100+)")
    if user_input[6] == 0:
        print("👉 Do 1 ML/AI project (GitHub)")
    if user_input[5] < 8:
        print("👉 Improve academic performance")

    # Next steps
    print("\n🚀 What should you do next:")
    print("👉 Practice English 2h/day")
    print("👉 Build 1 AI project")
    print("👉 Improve CV + GitHub portfolio")
```

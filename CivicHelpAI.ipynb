```python
# ---------- Импорт ----------
import re
from fpdf import FPDF
```

```python
# ---------- Функции анализа ----------
def smart_analysis(text):
    text_lower = text.lower()
    # Тип события и куда отправлять
    if "украл" in text_lower:
        event = "Кража"
        org = "Полиция"
    elif "дтп" in text_lower:
        event = "ДТП"
        org = "Полиция и скорая"
    elif "драка" in text_lower:
        event = "Насилие"
        org = "Полиция и скорая"
    elif "умер" in text_lower or "убил" in text_lower:
        event = "Смерть / травма"
        org = "Скорая и полиция"
    else:
        event = "Неопределено"
        org = "Полиция"

    urgency = "СРОЧНО 🚨" if any(w in text_lower for w in ["умер","кровь","убил"]) else "Обычное"
    time = re.findall(r"\d{1,2}:\d{2}", text)
    date = re.findall(r"\d{1,2}\.\d{1,2}\.\d{4}", text)
    place = "Не указано"
    if "в " in text_lower:
        place = text.split("в ")[-1]
    missing = []
    if not time: missing.append("время")
    if not date: missing.append("дата")
    if place == "Не указано": missing.append("место")

    return {"event": event, "urgency": urgency, "time": time, "date": date, "place": place, "missing": missing, "org": org}
```

```python
# ---------- Функция генерации заявления ----------
def generate_report_fast(user_input, analysis):
    date_text = analysis["date"][0] if analysis["date"] else "не указана"
    time_text = analysis["time"][0] if analysis["time"] else "не указано"
    place_text = analysis["place"]

    report = f"""
Куда: {analysis['org']}
От кого: Гражданин

Описание события:
{user_input}

Дата и время: {date_text} в {time_text}
Место: {place_text}

Просьба:
Прошу принять меры в соответствии с законом и обеспечить расследование данного события.
Срочность: {analysis['urgency']}
"""
    return report.strip()
```

```python
# ---------- Ввод события (вариант 1: текст прописан сразу) ----------
user_input = "Украли телефон в Алматы 24.03.2026 в 18:00"

analysis = smart_analysis(user_input)

print("\n📊 Анализ события:")
print(f"- Тип события: {analysis['event']}")
print(f"- Срочность: {analysis['urgency']}")
print(f"- Место: {analysis['place']}")
print(f"- Куда отправлять: {analysis['org']}")
if analysis["missing"]:
    print(f"⚠️ Не хватает данных: {', '.join(analysis['missing'])}")
else:
    print("✅ Данных достаточно")
```

```python
# ---------- Генерация заявления ----------
report = generate_report_fast(user_input, analysis)

print("\n===== Заявление =====\n")
print(report)
```

```python
# ---------- Сохранение в TXT ----------
with open("zayavlenie.txt", "w", encoding="utf-8") as f:
    f.write(report)
print("\n✅ Заявление сохранено в zayavlenie.txt")
```
